```python
import sys
sys.path.append('..')
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
from models.pde.Dimer import Dimer as model
```

```python
m = model(Da=0.1, Dp=0.1, konA=1, koffA=0.01, konP=1, koffP=0.01, kPA=0.1, kAP=0.1,
            alpha=1, beta=1, xsteps=100, psi=0.1, L=50, pA=1, pP=1, Tmax=100, deltat=0.01)
```

```python
m.initiate()
soln, time, solns, times = m.run(save_gap=1, kill_uni=True)
```

```python
import ipywidgets as widgets

def interactive_fig(solns, times):
    fig, ax = plt.subplots()
    ymax = 1.1 * max(np.max(solns[0]), np.max(solns[1]))
    @widgets.interact(time=(0, times[-1], 1))
    def update(time=0):
        ax.clear()
        tpoint = np.argmin(abs(times - time))
        ax.plot(solns[0][tpoint], label='aPAR')
        ax.plot(solns[1][tpoint], label='pPAR')
        ax.set_ylim(0, ymax)
        ax.set_xlabel('Position')
        ax.set_ylabel('Membrane concentration')
        ax.legend(loc='upper left')

    fig.set_size_inches(5,3)
    fig.tight_layout()
    
interactive_fig(solns, times)
```
